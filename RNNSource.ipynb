```python
#RNN - mfcc .wav to .npy

import os
import librosa
import numpy as np

# Define paths
base_path = '/content/drive/MyDrive/IRMASdataset'
input_path = os.path.join(base_path, 'ResampledDatasetFixed')
output_path = os.path.join(base_path, 'MFCCNpy')

instruments = ['cla', 'sax', 'flu', 'tru']
os.makedirs(output_path, exist_ok=True)

# Parameters
sample_rate = 22050
n_mfcc = 13

for instrument in instruments:
    instrument_input = os.path.join(input_path, instrument)
    instrument_output = os.path.join(output_path, instrument)
    os.makedirs(instrument_output, exist_ok=True)

    for filename in os.listdir(instrument_input):
        if filename.endswith('.wav'):
            filepath = os.path.join(instrument_input, filename)
            try:
                y, sr = librosa.load(filepath, sr=sample_rate)
                mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
                npy_filename = filename.replace('.wav', '.npy')
                np.save(os.path.join(instrument_output, npy_filename), mfcc)
                print(f"✅ Saved MFCC: {instrument}/{npy_filename}")
            except Exception as e:
                print(f"❌ Error processing {filename}: {e}")
```

```python
labels_dict = {"cla": 0, "sax": 1, "flu": 2, "tru": 3}
```

```python
#rnn - mel from .wav to .npy

import os
import librosa
import numpy as np


mel_output_path = os.path.join(base_path, 'MelSpectrogramNpy')  # base_path from previous step

os.makedirs(mel_output_path, exist_ok=True)

for instrument in instruments:
    instrument_input = os.path.join(input_path, instrument)
    instrument_output = os.path.join(mel_output_path, instrument)
    os.makedirs(instrument_output, exist_ok=True)

    for filename in os.listdir(instrument_input):
        if filename.endswith('.wav'):
            filepath = os.path.join(instrument_input, filename)
            try:
                y, sr = librosa.load(filepath, sr=sample_rate)
                mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
                npy_filename = filename.replace('.wav', '.npy')
                np.save(os.path.join(instrument_output, npy_filename), mel)
                print(f"✅ Saved Mel: {instrument}/{npy_filename}")
            except Exception as e:
                print(f"❌ Error processing {filename}: {e}")
```

```python
#RNN - chroma .wav to .npy

chroma_output_path = os.path.join(base_path, 'ChromaNpy')
os.makedirs(chroma_output_path, exist_ok=True)

for instrument in instruments:
    instrument_input = os.path.join(input_path, instrument)
    instrument_output = os.path.join(chroma_output_path, instrument)
    os.makedirs(instrument_output, exist_ok=True)

    for filename in os.listdir(instrument_input):
        if filename.endswith('.wav'):
            filepath = os.path.join(instrument_input, filename)
            try:
                y, sr = librosa.load(filepath, sr=sample_rate)
                chroma = librosa.feature.chroma_stft(y=y, sr=sr)
                npy_filename = filename.replace('.wav', '.npy')
                np.save(os.path.join(instrument_output, npy_filename), chroma)
                print(f"✅ Saved Chroma: {instrument}/{npy_filename}")
            except Exception as e:
                print(f"❌ Error processing {filename}: {e}")
```

```python
X_rnn = []
y_rnn = []

mel_dir = os.path.join(base_path, 'MelSpectrogramNpy')
mfcc_dir = os.path.join(base_path, 'MFCCNpy')
chroma_dir = os.path.join(base_path, 'ChromaNpy')

for instrument in instruments:
    mel_inst = os.path.join(mel_dir, instrument)
    mfcc_inst = os.path.join(mfcc_dir, instrument)
    chroma_inst = os.path.join(chroma_dir, instrument)

    filenames = [f for f in os.listdir(mel_inst) if f.endswith('.npy')]

    for file in filenames:
        try:
            mel = np.load(os.path.join(mel_inst, file)).T       # (130, 128)
            mfcc = np.load(os.path.join(mfcc_inst, file)).T     # (130, 13)
            chroma = np.load(os.path.join(chroma_inst, file)).T # (130, 12)

            # Confirm same time steps
            if mel.shape[0] == mfcc.shape[0] == chroma.shape[0] == 130:
                combined = np.concatenate((mel, mfcc, chroma), axis=1)  # (130, 153)
                X_rnn.append(combined)
                y_rnn.append(labels_dict[instrument])
            else:
                print(f"⚠️ Mismatch in {file}")
        except Exception as e:
            print(f"❌ Failed on {file}: {e}")

# Final arrays
X_rnn = np.array(X_rnn)
y_rnn = np.array(y_rnn)

print("✅ X_rnn shape:", X_rnn.shape)  # Expect (samples, 130, 153)
print("✅ y_rnn shape:", y_rnn.shape)
```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Bidirectional, Dropout, Dense

base_path = '/content/drive/MyDrive/IRMASdataset'

X_rnn = np.load(os.path.join(base_path, 'X_rnn.npy'))
y_rnn = np.load(os.path.join(base_path, 'y_rnn.npy'))
y_rnn_cat = to_categorical(y_rnn, num_classes=4)
```

```python
# === Apply 50% stratified subset to the RNN arrays ===
import numpy as np
import os
import pandas as pd

# Toggle here:
DATASET_PORTION = "50"   # "100" or "50"
SPLIT_SEED = 42

class_names = ['Clarinet', 'Saxophone', 'Flute', 'Trumpet']  # just for printouts

def print_class_counts(y_int, names):
    ids, cnts = np.unique(y_int, return_counts=True)
    for i, c in zip(ids, cnts):
        print(f"{names[i]}: {c} samples")

def stratified_half_indices(y_int, seed=42):
    rng = np.random.RandomState(seed)
    idx_sel = []
    for cls in np.unique(y_int):
        cls_idx = np.where(y_int == cls)[0]
        n = int(np.floor(len(cls_idx) * 0.5))
        chosen = rng.choice(cls_idx, size=n, replace=False)
        idx_sel.extend(chosen)
    return np.array(sorted(idx_sel), dtype=int)

print("🔍 Full dataset distribution:")
print_class_counts(y_rnn, class_names)

if DATASET_PORTION == "50":
    print(f"\n📉 Using stratified 50% of the dataset with seed = {SPLIT_SEED}")
    idx_sel = stratified_half_indices(y_rnn, seed=SPLIT_SEED)

    # Save manifest (for reproducibility)
    manifest_path = os.path.join(base_path, "subset_50_manifest_rnn.csv")
    pd.DataFrame({"index": idx_sel, "label": y_rnn[idx_sel]}).to_csv(manifest_path, index=False)
    print(f"📝 Saved subset manifest: {manifest_path}")

    # Apply subset
    X_rnn       = X_rnn[idx_sel]
    y_rnn       = y_rnn[idx_sel]
    y_rnn_cat   = y_rnn_cat[idx_sel]

    print("\n📊 Subset distribution:")
    print_class_counts(y_rnn, class_names)
    print("Shapes now -> X:", X_rnn.shape, " y:", y_rnn.shape, " y_cat:", y_rnn_cat.shape)
else:
    print("\n✅ Using 100% of the dataset (no reduction).")
```

```python
from tensorflow.keras.utils import to_categorical

y_rnn_cat = to_categorical(y_rnn, num_classes=4)
print(y_rnn_cat.shape)  # (1816, 4)
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_rnn, y_rnn_cat, test_size=0.2, stratify=y_rnn, random_state=42)

print("✅ X_train shape:", X_train.shape)
print("✅ X_test shape:", X_test.shape)
print("✅ y_train shape:", y_train.shape)
print("✅ y_test shape:", y_test.shape)
```

```python
from tensorflow.keras.layers import Input, BatchNormalization

def build_rnn_model_2(input_shape, num_classes):
    model2 = Sequential()
    model2.add(Input(shape=input_shape))
    model2.add(augment)

    # Deep stacked Bidirectional GRUs
    model2.add(Bidirectional(GRU(128, return_sequences=True)))
    model2.add(BatchNormalization())
    model2.add(Dropout(0.1))

    model2.add(Bidirectional(GRU(64, return_sequences=True)))
    model2.add(BatchNormalization())
    model2.add(Dropout(0.1))

    model2.add(Bidirectional(GRU(64)))
    model2.add(BatchNormalization())

    # Stronger Dense head
    model2.add(Dense(160, activation='relu'))
    model2.add(Dropout(0.1))
    model2.add(Dense(num_classes, activation='softmax'))

    return model2

# Match the optimizer style of CNN Model 2
opt = tf.keras.optimizers.AdamW(learning_rate=5e-4, weight_decay=5e-5)

model2 = build_rnn_model_2(input_shape, num_classes=4)
model2.compile(
    optimizer=opt,
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05),
    metrics=['accuracy']
)
```

```python
history2 = model2.fit(X_train, y_train, epochs=70, batch_size=16, validation_data=(X_test, y_test))
```

```python
# ==== METRICS, FIGURES, SAVING (single-label, multi-class) ====
# Requires: model, X_test, y_test, base_path

import os, json, numpy as np, matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    accuracy_score, f1_score, precision_score, recall_score,
    balanced_accuracy_score, cohen_kappa_score, matthews_corrcoef,
    average_precision_score, precision_recall_fscore_support
)

# --- identify this run ---
exp_id = "RNN_M2_50_holdout"   # e.g. CNN_M1_50_holdout, CNN_M2_100_holdout, ...

# --- paths ---
ts = datetime.now().strftime("%Y%m%d-%H%M%S")
out_dir = base_path
prefix = os.path.join(out_dir, f"{exp_id}_{ts}")

# --- predictions ---
y_true = y_test.argmax(axis=1)
y_prob = model2.predict(X_test, verbose=0)
y_pred = y_prob.argmax(axis=1)

class_names = ['Clarinet','Saxophone','Flute','Trumpet']
clarinet_idx = 0

# --- overall metrics ---
acc     = accuracy_score(y_true, y_pred)
prec_m  = precision_score(y_true, y_pred, average='macro',   zero_division=0)
prec_w  = precision_score(y_true, y_pred, average='weighted',zero_division=0)
rec_m   = recall_score(y_true, y_pred, average='macro',      zero_division=0)
rec_w   = recall_score(y_true, y_pred, average='weighted',   zero_division=0)
f1_m    = f1_score(y_true, y_pred, average='macro')
f1_w    = f1_score(y_true, y_pred, average='weighted')
bacc    = balanced_accuracy_score(y_true, y_pred)
kappa   = cohen_kappa_score(y_true, y_pred)
mcc     = matthews_corrcoef(y_true, y_pred)
top2    = (np.argsort(y_prob, axis=1)[:, -2:] == y_true[:, None]).any(axis=1).mean()

# --- per-class report ---
report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, digits=3)
report_text = classification_report(y_true, y_pred, target_names=class_names, digits=3)

# --- overall confusion matrix ---
cm = confusion_matrix(y_true, y_pred, labels=[0,1,2,3])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
fig, ax = plt.subplots(figsize=(5,5))
disp.plot(ax=ax, cmap='Blues', values_format='d', colorbar=False)
plt.title(f"Confusion Matrix — {exp_id}")
plt.tight_layout()
cm_path = f"{prefix}_cm.png"
plt.savefig(cm_path, dpi=300)
plt.close()

# --- one-vs-rest confusion matrices (each instrument) ---
ovr_paths = {}
for i, name in enumerate(class_names):
    y_true_bin = (y_true == i).astype(int)
    y_pred_bin = (y_pred == i).astype(int)
    cm_i = confusion_matrix(y_true_bin, y_pred_bin, labels=[0,1])
    disp_i = ConfusionMatrixDisplay(confusion_matrix=cm_i, display_labels=["Not Present","Present"])
    fig, ax = plt.subplots(figsize=(4,4))
    disp_i.plot(ax=ax, cmap='Purples', values_format='d', colorbar=False)
    plt.title(f"{name} OvR Confusion Matrix — {exp_id}")
    plt.tight_layout()
    pth = f"{prefix}_cm_{name.replace(' ','_')}_ovr.png"
    plt.savefig(pth, dpi=300)
    plt.close()
    ovr_paths[name] = pth

# --- clarinet-focused metrics (binary + PR-AUC) ---
y_true_c = (y_true == clarinet_idx).astype(int)
y_pred_c = (y_pred == clarinet_idx).astype(int)
prec_c, rec_c, f1_c, _ = precision_recall_fscore_support(y_true_c, y_pred_c, average='binary', zero_division=0)
pr_auc_c = average_precision_score(y_true_c, y_prob[:, clarinet_idx])

# --- bundle + save metrics ---
metrics = {
    "exp_id": exp_id,
    "timestamp": ts,
    "overall": {
        "accuracy": float(acc),
        "precision_macro": float(prec_m),
        "recall_macro": float(rec_m),
        "f1_macro": float(f1_m),
        "precision_weighted": float(prec_w),
        "recall_weighted": float(rec_w),
        "f1_weighted": float(f1_w),
        "balanced_accuracy": float(bacc),
        "cohen_kappa": float(kappa),
        "mcc": float(mcc),
        "top2_accuracy": float(top2),
    },
    "per_class_report": report_dict,
    "clarinet_specific": {
        "precision": float(prec_c),
        "recall": float(rec_c),
        "f1": float(f1_c),
        "pr_auc": float(pr_auc_c),
        "support": int(y_true_c.sum()),
    },
    "artifacts": {
        "cm_overall_png": cm_path,
        "cm_ovr_pngs": ovr_paths,
    }
}

json_path = f"{prefix}_metrics.json"
txt_path  = f"{prefix}_metrics.txt"
with open(json_path, "w") as f: json.dump(metrics, f, indent=2)
with open(txt_path, "w") as f:
    f.write(
        f"Experiment: {exp_id}\n"
        f"Accuracy: {acc:.3f}\n"
        f"Macro  — Precision: {prec_m:.3f}  Recall: {rec_m:.3f}  F1: {f1_m:.3f}\n"
        f"Weighted — Precision: {prec_w:.3f}  Recall: {rec_w:.3f}  F1: {f1_w:.3f}\n"
        f"Balanced Acc: {bacc:.3f} | Cohen's κ: {kappa:.3f} | MCC: {mcc:.3f}\n"
        f"Top-2 Acc: {top2:.3f}\n\n"
        "Per-class precision/recall/F1/support:\n"
        f"{report_text}\n\n"
        f"Clarinet — Precision: {prec_c:.3f}, Recall: {rec_c:.3f}, F1: {f1_c:.3f}, PR-AUC: {pr_auc_c:.3f}\n"
        f"Overall CM: {cm_path}\n"
        f"OvR CMs: {ovr_paths}\n"
    )

# --- save predictions for reproducibility ---
np.save(f"{prefix}_y_true.npy", y_true)
np.save(f"{prefix}_y_pred.npy", y_pred)
np.save(f"{prefix}_y_prob.npy", y_prob)

# --- save the trained model ---
model_path = f"{prefix}_model.keras"   # Keras native format
model.save(model_path)

print("Saved:")
print("  Metrics JSON:", json_path)
print("  Metrics TXT :", txt_path)
print("  Overall CM  :", cm_path)
for name, path in ovr_paths.items():
    print(f"  OvR CM ({name}): {path}")
print("  y_true/y_pred/y_prob .npy files")
print("  Model:", model_path)
```
